Download an image to apply vehicle detection to. The image is downloaded to the `content_dir` that we just created.

```python
!wget "https://yuanzsampledata.blob.core.windows.net/sample-data/car_traffic_moment.jpg?sv=2018-03-28&ss=b&srt=o&sp=r&se=2020-03-05T11:13:48Z&st=2019-08-05T03:13:48Z&spr=https&sig=7ye5uY4GL57FJPRALJdk%2FfyC0MVndhxB6AoYEGowf%2F4%3D" \
    -O content_dir/car_traffic_moment.jpg
```

```python
im = Image(filename="content_dir/car_traffic_moment.jpg", width=640)
display(im)
```

Apply vehicle detection using the vehicle detection script under `/scoring_app/detect_vehicle.py`.

- `--model-dir` - the model directory where all the model files are saved (optional)
- `--cuda` - 1 to use GPU, 0 to use CPU
- `--content-dir` - the directory of input images
- `--content-filename` the input image file name (optional)
- `--output-dir` - the directory to save output images

```python
!python scoring_app/detect_vehicle.py \
    --model-dir "models" \
    --cuda {1 if torch.cuda.is_available() else 0} \
    --content-dir "content_dir" \
    --content-filename "car_traffic_moment.jpg" \
    --output-dir "output_dir"
```

Check that there is something in the `output_dir`.

```python
!ls output_dir
```

Display the output image.

```python
im = Image(filename="output_dir/car_traffic_moment.jpg", width=640)
display(im)
```
